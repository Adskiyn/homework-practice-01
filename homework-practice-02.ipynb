```python
# Задача 1: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.
# Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.

import numpy as np
# Задаем матрицу
Nx=5
Ny=3
X=np.random.rand(Nx,Ny)*10
# Выводим матрицу
print('Исходная матрица генерируется случайным образом')
print(X)
# # Задача 1: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.
# Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.
# Создаем набор индексов по главной диагонали матрицы
di = np.diag_indices(np.min([Nx, Ny]))
#
print(di,X[di])
# Находим произведение элементов главной диагонали
res=np.prod(X[di])
print(res)
Ni=np.min(X.shape)
print(Ni)

# Оформляем все в виде функции
def task1_fun(X):
    # Определяем минимальный размер массива
    Ni=np.min(X.shape)
    # Создаем набор индексов по главной диагонали матрицы
    di = np.diag_indices(Ni)
    res=1.0
    for i in X[di]:
        if( i!=0 ):
            res=res*i
    return res

print(task1_fun(X))
```

```python
#Задача 2: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.
#Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ – True.
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
def task2_fun(xx, yy):
    # Сортируем элементы двух векторов
    xx.sort();
    yy.sort();
    # Возвращаем true если оба мультимножесмва совпадают
    if (xx == yy).all():
        return True
    else:
        return False
print(task2_fun(x,y))
```

```python
#Задача 3: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.
#Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ – 5.
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
def task3_fun(X):
    # Находим индексы всех нулевых элементов
#    idx=np.where(x==0)
    lx=X.size
    res=0
    for x,idx in enumerate(X):
        if ((x==0)&(idx<lx)):
            res=np.max([res,X[idx]])
    return res

print(task3_fun(x))
```

```python
#__ Задача 4__: Дан трёхмерный массив, содержащий изображение,
#размера (height, width, numChannels), а также вектор длины numChannels.
# Сложить каналы изображения с указанными весами, и вернуть результат в виде матрицы 
# размера (height, width). В ноутбуке приведите пример работы функции – преобразуйте  
#цветное изображение в оттенки серого, использовав коэффициенты np.array([0.299, 0.587, 0.114]).
# Считать реальное изображение можно при помощи функции scipy.misc.imread 
# (если изображение не в формате png, установите пакет pillow). 
# Обратите внимание, что в изображении может быть не три канала!
# За решения, явно или неявно использующие трёхканальность изображения будет 
# присуждена половина баллов.

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Загружаем картинку
# функции scipy.misc.imread в новых версиях нет
import imageio
#im = imageio.imread('elka.png')
im=mpimg.imread('elka.png')
print('Image size',im.shape)

to_gray=np.array([0.299, 0.587, 0.114])

def task4_fun(img,coefs):
    img1=np.zeros((img.shape[0],img.shape[1]),dtype=float)
    for i in range(0,img.shape[2]):
        img1[:,:]=img1[:,:]+coefs[i]*img[:,:,i]
    return img1

plt.imshow(im)
plt.title('Исходное изображение')
plt.show()
img1=task4_fun(im,to_gray)
plt.imshow(img1,cmap='gray')
plt.title('изображение')
plt.show()    
```

```python
#Задача 5: Реализовать кодирование длин серий (Run-length encoding). 
# Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. 
# Первый содержит числа, а второй - сколько раз их нужно повторить.
#Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) 
# ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).
x = np.array([2, 2, 2, 3, 3, 3, 5]) 

def task5_fun(x):
    qq=set(x)
    arr1=[]
    arr2=[]
    for q in qq:
        arr1.append(q)
        arr2.append(np.asarray(np.where(x==q)).size)
    return arr1,arr2
print(task5_fun(x))
```

```python
# Задача 6: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между 
# объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы 
# (сравнения приведите ниже в ноутбуке).
# За основу берем пример из документации scipy.spatial.distance.cdist
from scipy.spatial import distance
coords = [(35.0456, -85.2672),
          (35.1174, -89.9711),
          (35.9728, -83.9422),
          (36.1667, -86.7833)]
# 
print('Матрица расстояний через cdist')
print(distance.cdist(coords, coords, 'euclidean'))

def task6_fun(X,Y):
    lx=len(X)
    dists=np.zeros((lx,lx),dtype=float)
    
    for ix in range(0,lx):
        for iy in range(0,lx):
            dY=np.asarray(X[ix])-np.asarray(Y[iy])
            dists[ix,iy]=np.sqrt(np.dot(dY,dY))
    return dists
print('Матрица расстояний через нашу функцию')
print(task6_fun(coords,coords))
```

```python
def fit_Ridge(poly_deg,alphas):
    transf = PolynomialFeatures(degree=poly_deg)
# Дополняем количество признаков для обучающей выборки
    x_train_add = transf.fit_transform(x_train)
# Дополняем количество признаков для тестовой выборки
    x_test_add = transf.transform(x_test)
# Используем StandardScaler для выборок
    std1_scale_x = StandardScaler()
    std1_scale_y = StandardScaler()
    x1_std = std1_scale_x.fit_transform(x_train_add)
    y1_std = std1_scale_y.fit_transform(y_train.reshape(-1, 1)).flatten()
    x1_std_test=std1_scale_x.transform(x_test_add)
    y1_std_test = std1_scale_y.transform(y_test.reshape(-1, 1)).flatten()

    pipe_new1=Pipeline(steps=[('clf',Ridge())])
    
    param_grid=dict(clf__alpha=alphas)
    reg_new = GridSearchCV(pipe_new1, param_grid=param_grid)
    reg_new.fit(x1_std,y1_std)
#
    print('Best Ridge(StandardScaler). alpha=',reg_new.best_params_,'R2=',reg_new.score(x1_std,y1_std),'R2_test=',reg_new.score(x1_std_test,y1_std_test))
    return alphas,reg_new.cv_results_['mean_test_score'] 

#print(fit_Ridge(poly_deg,np.logspace(-6,6,13)))
#Xp,Yp=fit_Ridge(poly_deg,np.logspace(-6,6,50))  
#plt.semilogx(Xp,Yp)
```

```python
def fit_Lasso(poly_deg,alphas):
    transf = PolynomialFeatures(degree=poly_deg)
# Дополняем количество признаков для обучающей выборки
    x_train_add = transf.fit_transform(x_train)
# Дополняем количество признаков для тестовой выборки
    x_test_add = transf.transform(x_test)
# Используем StandardScaler для выборок
    std1_scale_x = StandardScaler()
    std1_scale_y = StandardScaler()
    x1_std = std1_scale_x.fit_transform(x_train_add)
    y1_std = std1_scale_y.fit_transform(y_train.reshape(-1, 1)).flatten()
    x1_std_test=std1_scale_x.transform(x_test_add)
    y1_std_test = std1_scale_y.transform(y_test.reshape(-1, 1)).flatten()

    pipe_new1=Pipeline(steps=[('clf',Lasso(max_iter=10000))])
    
    param_grid=dict(clf__alpha=alphas)
    reg_new = GridSearchCV(pipe_new1, param_grid=param_grid)
    reg_new.fit(x1_std,y1_std)
#
    print('Best Lasso(StandardScaler). alpha=',reg_new.best_params_,'R2=',reg_new.score(x1_std,y1_std),'R2_test=',reg_new.score(x1_std_test,y1_std_test))
    return alphas,reg_new.cv_results_['mean_test_score'] 
```
